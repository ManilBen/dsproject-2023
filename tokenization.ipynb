```python
import nltk
import spacy
import pandas as pd
import re

df = pd.read_csv("C:\\Users\\Utilisateur\\Desktop\\NLP NANCY\\YEAR 1\\Sem 2\\ds_project\\dsproject-2023\\segmentation_data.csv")

def tokenizer(df): # function to do tokenization part of the project
    unsegmented_text= str(df['original_text'][0]).lower() # the original text without sentence segmentation

    unsegmented_text_clean = re.sub(r'[\s+|\n+]', ' ', unsegmented_text).strip() # remove linebreaks and extra spaces from text before tokenizing

    model = spacy.load('en_core_web_sm')

    spacy_token_list = [token.text for token in model(unsegmented_text_clean)] # get all tokens in the text using spacy
    nltk_token_list = nltk.word_tokenize(unsegmented_text_clean) # get all tokens in the text using nltk

    set_u_token_nltk = set(nltk_token_list) # remove repeatitions by converting to set for nltk

    set_u_token_spacy = set(spacy_token_list) # remove repetitions by converting to set for spacy
    # print(set_u_token_spacy)


    size_vocab_nltk = len(set_u_token_nltk) # number of unique tokens identified by nltk in unsegmented text
    size_vocab_spacy = len(set_u_token_spacy) # number of unique tokens identified by spacy in unsegmented text

    

    common_tokens = set_u_token_nltk.intersection(set_u_token_spacy) # common tokens identified by both spacy and nltk libraries

    # print(size_vocab_nltk, size_vocab_spacy, len(common_tokens))

    unique_to_nltk_tokens = set_u_token_nltk.difference(common_tokens) # tokens uniquely identified by nltk library only
    unique_to_spacy_tokens = set_u_token_spacy.difference(common_tokens) # tokens uniquely identified by spacy library only

    
    segmented_token_list_spacy = []
    segmented_token_list_nltk = []
    segmented_token_list_spacy_notext = []

    for sentence in df['nltk_segmentation']: # loop to get tokens after sentence segmentation 
        segmented_text_clean = re.sub(r'[\s+|\n+]', ' ', str(sentence)).strip()

        segmented_token_list_spacy.extend(token.text for token in model(str(sentence)))
        
        segmented_token_list_spacy_notext.extend(token for token in model(str(sentence)))
        segmented_token_list_nltk.extend(nltk.word_tokenize(str(sentence)))

            
    return [segmented_token_list_nltk, segmented_token_list_spacy_notext] # returns the list of tokens in both libraries after sentence segmentation

    set_u_token_nltk_seg = set(segmented_token_list_nltk)
    set_u_token_spacy_seg = set(segmented_token_list_spacy)

    



tokenizer(df)
```
