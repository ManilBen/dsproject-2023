```python
import glob
import pandas as pd

# Set the directory path where the CSV files are stored
directory_path = 'E:\\Sports\\Archers'

# Get a list of all CSV files in the directory
csv_files = glob.glob(directory_path + '/*.csv')

# Loop through each CSV file and read it into a DataFrame
data_frames = []
for file in csv_files:
    df = pd.read_csv(file)
    data_frames.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(data_frames,ignore_index=True)

# Display the merged DataFrame
print(merged_df)
```

```python
#the newly concatenated information has index problem that is why we need to only catch columns we need into a new df
df_new = pd.DataFrame()
df_new['category'] = merged_df['category']
df_new['content'] = merged_df['content']
```
